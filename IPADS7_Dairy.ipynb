```
import pandas as pd
import statsmodels.api as sm
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
milk_df = pd.read_csv('drive/My Drive/Colab Notebooks/dairy.csv')
milk_df
```

```
milk_df.describe()
```

```
milk_data = milk_df.set_index(['FARM','YEAR'])
print(milk_data)
```

```
from linearmodels import PanelOLS

modmilk = PanelOLS(milk_data.MILK,milk_data[['LAND', 'COWS', 'LABOR', 'FEED']] )
fe_resmilk = modmilk.fit()
print(fe_resmilk)
```

```
from linearmodels.panel import RandomEffects
mod2milk = RandomEffects(milk_data.MILK,milk_data[['LAND', 'COWS', 'LABOR', 'FEED']] )
re_resmilk =mod2milk.fit()

print(re_resmilk)
```

```
re_resmilk.variance_decomposition
```

```
re_resmilk.theta.head()
```

```
from linearmodels.panel import BetweenOLS
mod_bols = BetweenOLS(milk_data.MILK,sm.add_constant(milk_data[['LAND', 'COWS', 'LABOR', 'FEED']] ))
be_res = mod_bols.fit()
print(be_res)
```

```
from linearmodels.panel import compare
print(compare({'FE':fe_resmilk,'RE':re_resmilk}))
```

```
from linearmodels import IV2SLS, IVLIML, IVGMM, IVGMMCUE

instrumentsmilk = ['COWS']
ivmodmilk = IV2SLS(milk_data.MILK, milk_data[[ 'LAND', 'FEED']], milk_data['LABOR'],milk_data[instrumentsmilk])
res2lsmilk = ivmodmilk.fit()
res2lsmilk.wu_hausman
```

```
from linearmodels.panel import FirstDifferenceOLS
mod_fd = FirstDifferenceOLS(milk_data.MILK,milk_data[['LAND', 'COWS', 'LABOR', 'FEED']])
fd_res = mod_fd.fit()
print(fd_res)
```
